## Importing Dataset

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition = fetch_ucirepo(id=544)

# data (as pandas dataframes)
X = estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.data.features
y = estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.data.targets

# metadata
print(estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.metadata)

# variable information
print(estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.variables)
```

```python
from pycaret.clustering import *
model = setup(X,verbose = False)
```

```python
NamesofModels = model.models().Name.index
```

```python
NamesofModels = NamesofModels[[0,2,4,5]]
```

```python
NamesofModels
```

```python
import pandas as pd
```

```python
searchSpace ={
    'No Preprocessing':{"remove_outliers": False,'transformation' : False,"normalize" :False ,'pca':False},
    'Transformation':{"remove_outliers": False,'transformation' : True,"normalize" :False ,'pca':False},
    'Normalisation':{"remove_outliers": False,'transformation' : False,"normalize" :True ,'pca':False},
    'PCA':{"remove_outliers": False,'transformation' : False,"normalize" :False ,'pca':True},
    'Norm + PCA':{"remove_outliers": False,'transformation' : False,"normalize" :True ,'pca':True},
    'Norm + PCA + TRANS':{"remove_outliers": False,'transformation' : True,"normalize" :True ,'pca':True},
}
```

```python
# Generating Tables
for mod in NamesofModels:
    rows = pd.DataFrame()
    for n_cluster in range(3, 6):
        for name,args in searchSpace.items():
          exp = setup(X, verbose=False,**args)
          create_model(mod, num_clusters=n_cluster,verbose = False)
          # print(len(rows))
          if len(rows) == 0:
              temp = exp.pull()
              temp['name'] = name
              temp['cluster_size'] = n_cluster
              rows = pd.DataFrame(temp)
          else:
              temp = exp.pull()
              temp['name'] = name
              temp['cluster_size'] = n_cluster
              rows = rows.append(temp)
    rows.set_index(['name','cluster_size'], inplace=True)
    rowsT =rows.sort_index().T
    rowsT.iloc[:3,:].to_csv(mod+'.csv')
    print(mod)
    display(rowsT.iloc[:3,:])
```
